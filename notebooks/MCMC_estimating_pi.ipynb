```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def estimate_pi(N):
    
    x = np.random.uniform(-1,1,N)
    y = np.random.uniform(-1,1,N)
    
    dist = np.sqrt(x**2 + y**2)
    
    ### Number of point in circle centered at origin with radius 1
    count = np.sum(dist <= 1.0)
    
    ### Estimate pi using ratio of areas of circle and square
    pi_estimate = 4 * count / N
    
    return pi_estimate
```

```python
N_array = np.array([10, 100, 1000, 10000, 100000, 1000000])

### vectorize function
v_estimate_pi = np.vectorize(estimate_pi)

### Use vectorized function to estimate pi for all N in N_array
pi_estimate_array = v_estimate_pi(N_array)

### Plotting
plt.loglog(N_array, np.abs(1 - pi_estimate_array / np.pi), marker='o')
plt.title("Relative error vs N")
plt.xlabel("N")
plt.ylabel("Relative error")
```
